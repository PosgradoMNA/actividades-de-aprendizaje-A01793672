##**Importar y Exportar Datos en Python**

**Importar Datos**

Es el proceso mediante el cual se cargan y leen datos dentro de Python desde varios recursos.

Para leer datos en Python usando la librerías de Pandas es importante considerar dos factores importantes:
* Formato:
  * .csv
  * .json
  * .xlsx
  * .hdf
* Ruta del archivo:
  * Computer
  * Internet

Sitio Web del dataset:
https://archive.ics.uci.edu/autos/imports-85.data

Al entrar al sitio, se puede apreciar como los datos están separados por comas, esto quiere decir que el formato es csv.

**Importando un CSV en Python**

```python
import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

df = pd.read_csv(url)
```

Con el método read_csv() se pueden importar los datos desde un csv y convertirlos a un DataFrame.
Esta instrucción asume que nuestro conjunto de datos tiene encabezados, en este caso no los tiene, por lo tanto, se debe especificar.

```python
df = pd.read_csv(url, header = None)
```

**Imprimir el DataFrame en Python**

* **df:** Imprime el dataframe completo (no se recomienda para conjuntos e datos largos).
* **df.head(n):** Para ver las primeras *n* filas del dataframe.
* **df.tail(n):** Muestra las últimas *n* filas del dataframe.

```python
df.head()
```

Como se puede apreciar, se le asignan números enteros a las columnas al no tenerse un encabezado como tal.

**Agregar Encabezado**

Ya que es difícil trabajar con el dataframe sin tener un significado de las columnas, se procede a asignarle nombres:

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style",
           "drive-wheels","engine-location","wheel-base","length","width","height","curb-weight",
           "engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","compression-ratio",
           "horsepower","peak-rpm","city-mpg","highway-mpg","price"]

df.columns=headers #Reemplaza el encabezado por defecto.
df.head()
```

**Exportar un DataFrame de Pandas a CSV**

```python
path="automobile.csv" #Ruta completa del archivo en tu computadora
df.to_csv(path)
```

En este curso solo se leeran y escribirán archivos csv pero Pandas soporta también otros tipos de archivos con diferente formato en los conjuntos de datos.

**Data Format** -- **Read** -- **Save**

csv -- pd.read_csv() -- df.to_csv()

json -- pd.read_json() -- df.to_json()

excel -- pd.read_excel() -- df.to_excel()

sql -- pd.read_sql() -- df.to_sql()

###**Conocimientos básicos de un Dataset**
**Tipos de datos**

Los datos tienen un variedad de tipos. Los principales datos que almacenan los objetos de Pandas son: objetos, flotantes, enteros y fechas.
El tipo objeto de Pandas funciona similar a un string.
Las fechas son usadas para el manejo de series a través del tiempo.

Es importante conocer el tipo de datos del dataset por los siguientes puntos:
* Discordancia entre el tipo de datos de una columna.
* Compatibilidad con los métodos de Python.

```python
df.dtypes #Despliega una lista con la relación del tipo de dato de cada columna
```

Se debe verificar que el tipo de dato concuerde con la descripción de la columna.

Por ejemplo, la columna "bore" tiene un tipo de dato incorrecto, ya que se refiere a una dimensión del motor, por lo tanto se espera un dato de tipo numérico. Esto se debe corregir.

Si se desea conocer un resumen estadístico de cada columna, se aplica lo siguiente:

```python
df.describe()
```

El método *describe()* omite las filas y columnas que no contienen datos numéricos.

Aún así es posible aplicar este método a datos del tipo objeto. Para aplicar esta función a todas las columnas se ejecuta el siguiente argumento:

```python
df.describe(include="all")
```

Se puede apreciar que para las columnas de tipo objeto solo se hacen validas tres estadísticas:
* **Unique:** número de los diferentes objetos en la columna.
* **Top:** es similar a la moda.
* **Freq:** se refiere al número de veces que aparece el objeto en el top.

**NaN:** Siglas para Not a Number.
